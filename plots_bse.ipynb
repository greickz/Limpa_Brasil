```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("planilha_GERAL.xlsx")
```

```python
plt.figure()
df["Campanha"].value_counts().plot(kind="bar", color="slateblue")
plt.title("Frequência de Cada Campanha")
plt.xlabel("Campanha")
plt.ylabel("Número de Registros")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Substitua pelo caminho correto do arquivo
arquivo = "planilha_GERAL.xlsx"
df = pd.read_excel(arquivo, sheet_name="DF LIMPO")

# Ajustar data
df["Data"] = pd.to_datetime(df["Data"], errors="coerce")
df["AnoMes"] = df["Data"].dt.to_period("M")
df.head()
```

```python
df["Ano"] = df["Data"].dt.year
df["Ano"].value_counts().sort_index().plot(kind="bar", color="lightsteelblue", figsize=(8, 5))
plt.title("Distribuição de Campanhas por Ano")
plt.xlabel("Ano")
plt.ylabel("Número de Campanhas")
plt.tight_layout()
plt.show()
```
